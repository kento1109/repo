## sin波の生成(RGAN)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# set hyper parameter for genearating data
samples = 10000
sequence = 30
```

```python
def make_sin(T, noise=False, ampl=0.1, freq=1.0):
    x = np.arange(-np.pi * 2, np.pi * 2, 0.25)
    start_seq = np.random.randint(0, x.shape[0] - T, 1)[0]
    sin = np.sin(x[start_seq: start_seq + T] * freq) * 0.5
    noise_ = ampl * np.random.uniform(low=-1.0, high=1.0, size=len(sin))
    return sin + noise_ if noise else sin
```

```python
# generate training dataset
sin_x = []
for _ in range(samples):
    ampl = np.random.uniform(low=0.01, high=0.1)
    freq = np.random.uniform(low=1.0, high=2.5)
    sin_x.append(make_sin(sequence, noise=False, ampl=ampl, freq=freq))
sin_x = np.array(sin_x).reshape(samples, sequence, 1)

print('train x shape: ', sin_x.shape)

sin_samples = 15
idxes = np.random.randint(0, sin_x.shape[0], sin_samples)
plt.figure(figsize=(20, 5))
# show sin wave
for i in range(sin_samples):
    plt.subplot(3, sin_samples / 3, i + 1)
    plt.plot(sin_x[idxes[i]].squeeze())
    plt.axis('off')
    plt.ylim(-0.8, 0.8)
```

```python
# split train data
X_train, X_valid = train_test_split(sin_x, test_size=0.1, random_state=0)
train_size = X_train.shape[0]
valid_size = X_valid.shape[0]
print('train shape:', X_train.shape)
print('valid shape:', X_valid.shape)
```

```python
from keras.layers import Input, Dense, Reshape, Flatten, Dropout, RepeatVector
from keras.layers import BatchNormalization, Activation, ZeroPadding2D
from keras.layers.recurrent import LSTM
from keras.models import Sequential, Model
from keras.optimizers import RMSprop
```

```python
class RGAN():
    def __init__(self, 
                 batch_size,
                 input_dim, 
                 latent_dim, 
                 hidden_dim, 
                 sequence):
        self.batch_size = batch_size
        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.hidden_dim = hidden_dim
        self.sequence = sequence

        optimizer = RMSprop()

        # Build and compile the discriminator
        self.discriminator = self.build_discriminator()
        self.discriminator.compile(loss='binary_crossentropy',
            optimizer=optimizer,
            metrics=['accuracy'])

        # Build the generator
        self.generator = self.build_generator()

        # The generator takes noise as input and generates imgs
        z = Input(shape=(self.sequence, self.latent_dim))
        gen = self.generator(z)

        # For the combined model we will only train the generator
        self.discriminator.trainable = False

        # The discriminator takes generated images as input and determines validity
        validity = self.discriminator(gen)

        # The combined model  (stacked generator and discriminator)
        # Trains the generator to fool the discriminator
        self.combined = Model(z, validity)
        self.combined.compile(loss='binary_crossentropy', optimizer=optimizer)


    def build_generator(self):

        model = Sequential()

        model.add(LSTM(units=self.hidden_dim, 
                       batch_input_shape=(None, self.sequence, self.latent_dim), 
                       return_sequences=True))
        model.add(Dense(self.input_dim, activation='tanh'))
#         model.summary()

        noise = Input(shape=(self.sequence, self.latent_dim))
        gen_x = model(noise)

        return Model(noise, gen_x)

    def build_discriminator(self):

        model = Sequential()

        model.add(LSTM(units=self.hidden_dim, 
                       batch_input_shape=(None, self.sequence, self.input_dim), 
                       return_sequences=True))
        model.add(Dense(1, activation='sigmoid'))
#         model.summary()

        inputs = Input(shape=(self.sequence, self.input_dim))
        validity = model(inputs)

        return Model(inputs, validity)

    def train(self, n_epochs, x_train):

        # Adversarial ground truths
        valid = np.ones((self.batch_size, self.sequence, self.input_dim))
        fake = np.zeros((self.batch_size, self.sequence, self.input_dim))

        for epoch in range(n_epochs):
            
            np.random.shuffle(x_train)
            
            # ---------------------
            #  Train Discriminator
            # ---------------------
            
            for i in range(int(sin_x.shape[0] / batch_size)):
                
                # Select a random batch of 
                x = x_train[i*batch_size:(i+1)*batch_size]

                noise = np.random.normal(0, 1, 
                                         (self.batch_size, 
                                          self.sequence, 
                                          self.latent_dim))

                # Generate a batch of new images
                gen_x = self.generator.predict(noise)

                # Train the discriminator
                d_loss_real = self.discriminator.train_on_batch(x, valid)
                d_loss_fake = self.discriminator.train_on_batch(gen_x, fake)
                d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

                # ---------------------
                #  Train Generator
                # ---------------------

                noise = np.random.normal(0, 1, 
                                         (self.batch_size, 
                                          self.sequence, 
                                          self.latent_dim))

                # Train the generator (to have the discriminator label samples as valid)
                g_loss = self.combined.train_on_batch(noise, valid)

            # Plot the progress
            print ("%d [D loss: %f, acc.: %.2f%%] [G loss: %f]" 
                   % (epoch + 1, d_loss[0], 100*d_loss[1], g_loss))
```

```python
# hyper parameter for training
input_dim = sin_x.shape[2]
latent_dim = 50
hidden_dim = 100
batch_size = 64
n_epochs = 50

rgan = RGAN(batch_size, 
            input_dim, 
            latent_dim, 
            hidden_dim, 
            sequence)
```

```python
rgan.train(n_epochs, sin_x)
```

```python
def generate_sin(generator, sample_size=20):
    noise = np.random.normal(0, 1, (sample_size, sequence, latent_dim))
    gen_sin = generator.predict(noise)
    return np.squeeze(gen_sin)
```

## 未学習のモデルでのsin波復元
---

```python
sin_samples = 20

rgan_ = RGAN(batch_size, 
            input_dim, 
            latent_dim, 
            hidden_dim, 
            sequence)

gen_x_sample = generate_sin(rgan_.generator)

plt.figure(figsize=(20, 12))
# show sin wave
for i in range(sin_samples):
    plt.subplot(5, 4, i + 1)
    plt.plot(gen_x_sample[i].squeeze())
    plt.axis('off')
    plt.ylim(-0.8, 0.8)
```
